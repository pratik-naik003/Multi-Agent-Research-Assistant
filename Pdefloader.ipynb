```python
# ================================
# Install Dependencies
# ================================
!pip install -q langchain langchain-community pypdf

# ================================
# Upload PDF
# ================================
from google.colab import files
uploaded = files.upload()

# Get uploaded file name
pdf_path = list(uploaded.keys())[0]
print("Uploaded PDF:", pdf_path)

# ================================
# Load PDF using PyPDFLoader
# ================================
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader(pdf_path)
documents = loader.load()

print(f"\nTotal pages loaded: {len(documents)}")

# ================================
# Show first page content
# ================================
print("\n--- First Page Content ---\n")
print(documents[0].page_content[:1000])

# ================================
# Split Text (Optional but useful for RAG)
# ================================
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)

chunks = text_splitter.split_documents(documents)

print(f"\nTotal text chunks created: {len(chunks)}")

# ================================
# Show sample chunk
# ================================
print("\n--- Sample Chunk ---\n")
print(chunks[0].page_content[:500])
```
